Perform EDA (Exploratory Data Analysis) on COVID-19 data using Pandas, Matplotlib, and Seaborn.

**Introduction**

The COVID-19 pandemic has significantly impacted global health, economics, and daily life. In this project, we performed Exploratory Data Analysis (EDA) on global COVID-19 data to gain insights into case trends, death counts, and recovery statistics across different countries. The dataset used includes key attributes such as confirmed cases, deaths, recoveries, and country-wise progression over time.

We used Python with libraries like Pandas, Matplotlib, and Seaborn to clean, analyze, and visualize the data. Our focus was not on prediction but on extracting meaningful patterns and presenting them through intuitive charts.

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

Load the Dataset

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('covid_19_clean_complete.csv')
df.head()
```

Clean the Data

```python
# Rename columns to standard lowercase names
df = df.rename(columns={
    'Date': 'date',
    'Country/Region': 'country',
    'Confirmed': 'total_cases',
    'Deaths': 'total_deaths',
    'Recovered': 'total_recovered',
    'Active': 'active_cases'
})

# Keep only useful columns
df = df[['date', 'country', 'total_cases', 'total_deaths', 'total_recovered', 'active_cases']]

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Sort data for new case calculation
df = df.sort_values(by=['country', 'date'])

# Calculate new cases
df['new_cases'] = df.groupby('country')['total_cases'].diff().fillna(0)

df.head()

```

Top 10 Countries by Total Cases

```python
latest_date = df['date'].max()
top_countries = df[df['date'] == latest_date].groupby('country')['total_cases'].sum().sort_values(ascending=False).head(10)

top_countries.plot(kind='bar', figsize=(10,6), title='Top 10 Countries by Total Cases')
plt.ylabel('Total Cases')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 New Cases Trend Over Time ( India, US, Brazil)

```python
countries = ['India', 'United States', 'Brazil']
for country in countries:
    subset = df[df['country'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.title('New COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

Bar Chart: Total Deaths vs Total Recoveries (if recoveries column exists)

```python
latest = df[df['date'] == latest_date]

recoveries_deaths = (
    latest.groupby('country')[['total_deaths', 'total_recovered']]
    .sum()
    .dropna()
    .sort_values(by='total_deaths', ascending=False)
    .head(10)
)

recoveries_deaths.plot(kind='bar', figsize=(10,6))
plt.title('Total Deaths vs Recoveries (Top 10 Countries)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

Group Data by Date or Country



```python
global_trend = df.groupby('date')['total_cases'].sum()
global_trend.plot(figsize=(10,6), title='Global COVID-19 Case Growth')
plt.ylabel('Total Cases')
plt.xlabel('Date')
plt.show()
```

Line Graph for India's Case Growth

```python
india_data = df[df['country'] == 'India']
plt.plot(india_data['date'], india_data['total_cases'], label='India')
plt.title('India COVID-19 Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.show()
```

Calculate Death Rate

```python
df['death_rate'] = (df['total_deaths'] / df['total_cases']) * 100
df[['country', 'date', 'death_rate']].tail()
```
